# selection using iloc and loc

####  df.iloc: index based selection

- select single row and single column 
- select single row and multiple column 
- select multiple rows and multiple columns

```python
market.head()
```

```python
market.iloc[0:6, 3:8]          #first the row numbers then the column numbers
```

```python
market.iloc[2,4]
```

```python
# .iloc[[list of rows you want -non adjacent],[list of non adjacent cols]]

#iloc treats the row index numbers as 'labels', like a name;; 

market.iloc[[2,6,8],[3,4,7]]
```

 ### df.loc is a label based selection

```python
 market.loc[7:14, 'Cust_id':'Profit']
```

```python
market.loc[0:5,['Cust_id', 'Sales','Profit']]
```

#### tools to get index numbers of column
- get_loc() - index no. of a single columns
- get_indexer() - index no. of multiple columns

```python
market.columns.get_loc('Profit')
```

```python
market.columns.get_indexer(['Profit','Sales'])
```

## Merging the DataFrames


```python
market = pd.read_csv('market_fact.csv')
customer = pd.read_csv('cust_dimen.csv')
product = pd.read_csv('prod_dimen.csv')
```

```python
market.head()
```

```python
customer.head()
```

```python
product.head()
```

```python
new_df= pd.merge(market, customer, how = 'inner', on = 'Cust_id')
```

```python
new_df
```

 if key names are different- pd.merge(market, customer, how = 'inner', left_on = 'Cust_id', right_on = 'ID')
- left data frame has the col cust_id and the name of the same in right df is 'id'

```python
merged_data = pd.merge(new_df, product, how = 'inner', on = 'Prod_id')
```

```python
merged_data
```

```python
#exporting merged data as a csv file

merged_data.to_csv('Merged_DataFrames.csv')

# it will save with index numbers in the first column
```

```python
merged_data.to_csv('Merged_DataFrames2.csv', index=False)
#it will save without index numbers
```
