# Concatenating the Data
concatenation is much more straightforward than merging. It is used when you have dataframes having the same columns and want to append them (pile on top of other) or having the same rows and want to append them side by side. 

### Concatenating DataFrames having the Same Columns
say you have two data frames having the same columns, like so:
-when you have same no. of columns and columns are named the same

```python
df1 = pd.DataFrame({'Name': ['Aman', 'Joy', 'Rashmi', 'Saif'],
                    'Age': ['34', '31', '22', '33'],
                    'Gender': ['M', 'M', 'F', 'M']})
```

```python
df2 = pd.DataFrame({'Name': ['Akhil', 'Asha', 'Preeti'],
                    'Age': ['31', '22', '23'],
                    'Gender': ['M', 'F', 'F']})
```

```python
pd.concat([df1,df2], axis = 0) #you can concat any no. of df, unlike merge in pd
```

```python
pd.concat([df1,df2], axis = 0, ignore_index = True)  #do ignore_index to rest the index numbers
```

### Concatenating DataFrames having the Same Rows
say you have two data frames having the same rows, like so:
-when you have diff no. of columns and columns are named the different but no. of rows is same

```python
 df3 = pd.DataFrame({'Name': ['Aman', 'Joy', 'Rashmi', 'Saif'],
                    'Age': ['34', '31', '22', '33'],
                    'Gender': ['M', 'M', 'F', 'M']})
```

```python
df4 = pd.DataFrame({'School': ['RK Public', 'JSP', 'Carmel Convent', 'St. Paul'],
                    'Graduation Marks': ['84', '89', '76', '91']})
```

```python
pd.concat([df3,df4], axis = 1)
```

# Grouping and Summarizing the data

```python
data = pd.read_csv('Merged_DataFrames2.csv')
```

```python
data.head()
```

```python
data['Profit'].sum()
```

```python
data['Product_Category'].unique() #works only on a categorical column, not for numeric column
```

```python
by_category = data.groupby('Product_Category')  #creating a variable to hold the groups
```

```python
by_category.groups
```

```python
by_category.get_group('TECHNOLOGY')     #case sensitive, write as its written
```

```python
by_category["Profit"].sum()
```

```python
##grouping 2 columns can happen when the categories are related
by_subcat= data.groupby(['Product_Category', 'Product_Sub_Category'])

    #use a list when writing multiple columns
```

```python
by_subcat['Profit'].sum()
```
